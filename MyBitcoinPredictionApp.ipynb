### Part 1 - DATAFRAME, CLEAN AND PREPARE DATA.

```python
# Frist, Import Lib needed for our Bitcoin prediction and ETL.
import pandas as pd
import numpy as np
import time
import datetime
import pandas as pd
import numpy as np

# Import Data and Convert into DF
df = pd.read_csv("./Resources/Bitcoin_Historical_Data/Bitcoin_1_Min_Historical_Data_2012_to_2020.csv")
# show the first 5 rows
df.head()
```

```python
# Convert Timestamp into TimeZone (ET)
df['Timestamp'] = (pd.to_datetime(df['Timestamp'], unit='s')
                     .dt.tz_localize('est')
                     .dt.tz_convert('America/New_York'))
# show the first 5 rows
df.head()
```

```python
# show the last 5 rows
df.tail()
```

```python
#Drop NaNs from Dataset. 
df.dropna(inplace=True)
# show the first 5 rows
df.head()
```

```python
# Clean DataFrame for Prediction utilization. 
# Remove Volume Currency from DF
df.drop(['Volume_(Currency)'],1,inplace=True)
# show the first 5 rows
df.head()
```

```python
# Clean DataFrame for Prediction utilization. 
# Remove Volume Currency from DF
df.drop(['Volume_(BTC)'],1,inplace=True)
# show the first 5 rows
df.head()
```

```python
# Rename Timestamp and Weighted_Price
df.rename(columns = {'Timestamp':'Last Update', 'Weighted_Price':'Weighted Price'}, inplace = True)
# show the first 5 rows
df.head()
```

```python
# Clean Last Update into ms.
df['Last Update'] = df['Last Update'].values.astype(dtype='datetime64[ms]')
# show the first 5 rows
df.head()
```

```python
# Esport Clean DF into CSV
df.to_csv (r'./Resources/Bitcoin_Historical_Data/Bitcoin_2011-2020_Clean.csv', index = False, header=True)
```

```python
# Let's create a new DF to use for our prediction, removing the Last Update (Timestamp)
wp_df = df.copy(deep=True)
wp_df.head()
```

### Let's use 2020 Bitcoin Data only.

```python
#Let's filter new DataFrame for 2020 year/data only. 
wp2020_df = wp_df[(wp_df["Last Update"] >= '2020-10-01')]
wp2020_df.head()
```

```python
wp2020_df.tail()
```

```python
# Esport Clean 2020 DF into CSV
wp2020_df.to_csv (r'./Resources/Bitcoin_Historical_Data/Bitcoin_Q32020_Clean.csv', index = False, header=True)
```

### Let's Clean the 2020 Columns and Data

```python
# Remove Volume Currency from DF
wp2020_df.drop(['Last Update'],1,inplace=True)
# show the first 5 rows
wp2020_df.head()
```

```python
# Remove Volume Currency from DF
wp2020_df.drop(['Open'],1,inplace=True)
# show the first 5 rows
wp2020_df.head()
```

```python
# Remove Volume Currency from DF
wp2020_df.drop(['High'],1,inplace=True)
# show the first 5 rows
wp2020_df.head()
```

### Part 2 - PREPARE DATA FOR BITCOIN PRICE PREDICTION

#### 30 Days Prediction Model

```python
# Machine Learning Coding. 
# 30 Days Prediction. 
```

```python
# Create Variable bitCoinPredictionDays to set the days.
bitCoinPredictionDays = 30
```

```python
# Create another column shifted 'n'  units up
wp2020_df['Prediction > 30 Days'] = wp2020_df[['Weighted Price']].shift(-bitCoinPredictionDays)
# show the first 5 rows
wp2020_df.head()
```

```python
# show the last 5 rows from new dataframe
wp2020_df.to_csv (r'./Resources/Bitcoin_Historical_Data/30_Days_Price_Prediction_Bitcoin_Q32020_Clean.csv', index = False, header=True)
wp2020_df.tail()
```

### Part 3 - CREATE AN INDEPENDENT DATASET

```python
# Let's Create the independent Dataset.
# Here we will convert our wp_df Dataframe into a Numpy Array and drop the "Prediction > 30 Days" column
x = np.array(wp2020_df.drop(['Prediction > 30 Days'],1))

# Remove the last 'n' rows where 'n' is the bitCoinPredictionDays
x = x[:len(wp2020_df)-bitCoinPredictionDays]
print(x)
```

### Part 4 - CREATE A DEPENDENT DATASET

```python
# Create the dependent Dataset.
# Convert the DataFrame into a Numpy Array
y = np.array(wp2020_df['Prediction > 30 Days'])

# Get all the values except last 'n' rows
y = y[:-bitCoinPredictionDays]
print(y)
```

### Part 5 - SPLIT DATA FOR MACHINE LEARNING TEST

```python
# Import libs, and Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2)
```

```python
# Set the predictionDays array equal to last 30 rows from the original data set
bitCoinPredictionDays_array = np.array(wp2020_df.drop(['Prediction > 30 Days'],1))[-bitCoinPredictionDays:]

print(bitCoinPredictionDays_array)
```

## Now let's create a Machine Learning Model

### Part 6 - CREATE AND TRAIN VECTOR MACHINE MODEL 

```python
# Start Vector Machine (Regression)
```

```python
# We need first to install sklearn if missing: (We already installed - as comment below)
#pip install sklearn
```

```python
from sklearn.svm import SVR
# Create and Train the Support Vector Machine (Regression) using radial basis function
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain, ytrain)
```

## Now let's Test the Machine Learning Model. 

### Part 7 - TEST MODEL

```python
# Create a test.
svr_rbf_confidence = svr_rbf.score(xtest,ytest)
print('SVR_RBF accuracy :',svr_rbf_confidence)
```

```python
# print the predicted values
svm_prediction = svr_rbf.predict(xtest)
print(svm_prediction)
print()
print(ytest)
```
